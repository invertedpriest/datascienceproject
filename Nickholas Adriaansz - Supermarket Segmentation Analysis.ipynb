# **Feature Engineering**

## Adding New Column

**New Column: 'Age'**  
Assuming the dataset reflects data as of 2021, the age is calculated accordingly.

```python
df1 = df.copy()
```

```python
#Insert 'Age' column
refYear = 2021
df1.insert(2, 'Age', value=refYear - df1['YearOfBirth'])
df1
```

```python
df1['Age'].describe()
```

**New Column: TotalChildren**  
Inserting TotalChildren column to reflect the number of children living in the same house.

```python
#Insert TotalChildren column after Teenhome's column
df1.insert(8, 'TotalChildren', value=df1['Kidhome'] + df1['Teenhome'])
df1
```

**New Column: TotalMntSpent**  
To calculate customers' spending across all product categories:
- Wines
- Fruits
- Meat products
- Fish products
- Sweet products
- Gold products

```python
#Insert TotalMntSpent column
TotalMntSpent = ['MntWines', 'MntFruits', 'MntMeatProducts', 
               'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']

df1.insert(17, 'TotalMntSpent', value=df1[TotalMntSpent].sum(axis=1))
df1
```

**New Column: EnrollmentYear**  
To calculate how long a customer has been with the company.

```python
# Insert EnrollmentYear
df1.insert(10, 'EnrollmentYear', df1['CustEnrollmentDate'].dt.year)
df1
```

**New Column: CustomerTenure**  
To calculate number of months a customer has been with the company.

```python
# Set reference date
refDate = pd.to_datetime('2021-12-31')

# Calculate CustomerTenure in months and insert at index 11
df1.insert(11, 'CustomerTenure', 
           ((refDate - df1['CustEnrollmentDate']).dt.days // 30).astype(int))
df1
```

# **Recommendation**

## Target No Children Customers

```python
#Basic calculations from actual data
totalCustomers = len(df)
currentTotalRevenue = df1['TotalMntSpent'].sum()
additionalRevenueNeeded = currentTotalRevenue * 0.10

print("BASELINE DATA FROM DATASET:")
print("Total customers:", totalCustomers)
print("Current total revenue: $", currentTotalRevenue)
print("Additional revenue needed: $", additionalRevenueNeeded)
```

```python
#Filter No Children customers from dataset
noChildrenSegment = df1[df1['TotalChildren'] == 0]

#Calculate current performance
noChildrenCustomers = len(noChildrenSegment)
noChildrenCurrentTotal = noChildrenSegment['TotalMntSpent'].sum()
noChildrenCurrentAvg = noChildrenCurrentTotal / noChildrenCustomers

print("NO CHILDREN SEGMENT - CURRENT:")
print("Customers:", noChildrenCustomers)
print("Current total revenue: $", noChildrenCurrentTotal)
print("Current avg spending: $", noChildrenCurrentAvg)
```

```python
#Calculate 12% increase impact
growthRate = 0.12
newAvgSpending = noChildrenCurrentAvg * (1 + growthRate)
newTotalRevenue = noChildrenCustomers * newAvgSpending
additionalRevenue = newTotalRevenue - noChildrenCurrentTotal

print("NO CHILDREN RECOMMENDATION IMPACT:")
print("Growth rate: 12%")
print("New avg spending: $", round(newAvgSpending, 0))
print("New total revenue: $", round(newTotalRevenue, 0))
print("Additional revenue: $", round(additionalRevenue, 0))

#Calculate percentage of target achieved
percentOfTarget = (additionalRevenue / additionalRevenueNeeded) * 100
print("Percentage of 10% target achieved:", round(percentOfTarget, 1), "%")
```

## Focus on High-Income Customers

```python
#Filter High Income customers from dataset  
highIncomeSegment = df1[df1['IncomeGroup'] == 'High Income']

#Calculate current performance
highIncomeCustomers = len(highIncomeSegment)
highIncomeCurrentTotal = highIncomeSegment['TotalMntSpent'].sum()
highIncomeCurrentAvg = highIncomeCurrentTotal / highIncomeCustomers

print("HIGH INCOME SEGMENT - CURRENT:")
print("Customers:", highIncomeCustomers)
print("Current total revenue: $", highIncomeCurrentTotal)
print("Current avg spending: $", round(highIncomeCurrentAvg, 0))
```

```python
#Calculate 15% increase impact  
highIncomeGrowthRate = 0.15
highIncomeNewAvg = highIncomeCurrentAvg * (1 + highIncomeGrowthRate)
highIncomeNewTotal = highIncomeCustomers * highIncomeNewAvg
highIncomeAdditionalRevenue = highIncomeNewTotal - highIncomeCurrentTotal

print("HIGH INCOME RECOMMENDATION IMPACT:")
print("Growth rate: 15%")
print("New avg spending: $", round(highIncomeNewAvg, 0))
print("New total revenue: $", round(highIncomeNewTotal, 0))
print("Additional revenue: $", round(highIncomeAdditionalRevenue, 0))

#Calculate percentage of target achieved
highIncomePercentOfTarget = (highIncomeAdditionalRevenue / additionalRevenueNeeded) * 100
print("Percentage of 10% target achieved:", round(highIncomePercentOfTarget, 1), "%")
```

```python
#Combine No Children + High Income impact
totalAdditionalRevenue = additionalRevenue + highIncomeAdditionalRevenue
totalPercentOfTarget = (totalAdditionalRevenue / additionalRevenueNeeded) * 100

print("COMBINED RECOMMENDATIONS IMPACT:")
print("No Children additional revenue: $", round(additionalRevenue, 0))
print("High Income additional revenue: $", round(highIncomeAdditionalRevenue, 0))
print("Total additional revenue: $", round(totalAdditionalRevenue, 0))
print("Total target needed: $", round(additionalRevenueNeeded, 0))
print("Total percentage achieved:", round(totalPercentOfTarget, 1), "%")

#Check if target is achievable
if totalAdditionalRevenue >= additionalRevenueNeeded:
    print("TARGET ACHIEVED!")
else:
    shortfall = additionalRevenueNeeded - totalAdditionalRevenue
    print("Still need: $", round(shortfall, 0), "more revenue")
```
